```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
experiments = pd.read_pickle('/home/student/Downloads/SimPLE_experiments.pkl')
```

## Pong

```python
pong_exp = experiments[experiments['game']== 'pong']
```

```python
pong_exp.head()
```

```python
pong_exp = pong_exp.drop('experiment_id', 1)
pong_exp = pong_exp.drop('experiment_name', 1)
pong_exp = pong_exp.drop('game', 1)
```

```python
plt.figure(figsize=(25,15))
for i in range(5):
    index = 141 + i
    pong_exp_allepoch = experiments[(experiments['game']== 'pong') &
                           (experiments['experiment_id'] == index)][-15:]
    plt.subplot(5,3,(i+1))
    plt.axhline(y=0, color='black')
    
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.0_max_noops_0_clipped'], label='temp_0.0_max_noops_0_clipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.0_max_noops_0_unclipped'], label='temp_0.0_max_noops_0_unclipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.0_max_noops_8_clipped'], label='temp_0.0_max_noops_8_clipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.0_max_noops_8_unclipped'], label='temp_0.0_max_noops_8_unclipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.5_max_noops_0_clipped'], label='temp_0.5_max_noops_0_clipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.5_max_noops_0_unclipped'], label='temp_0.5_max_noops_0_unclipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.5_max_noops_8_clipped'], label='temp_0.5_max_noops_8_clipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_0.5_max_noops_8_unclipped'], label='temp_0.5_max_noops_8_unclipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_1.0_max_noops_0_clipped'], label='temp_1.0_max_noops_0_clipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_1.0_max_noops_0_unclipped'], label='temp_1.0_max_noops_0_unclipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_1.0_max_noops_8_clipped'], label='temp_1.0_max_noops_8_clipped')
    plt.plot(pong_exp_allepoch['epoch_num'], pong_exp_allepoch['temp_1.0_max_noops_8_unclipped'], label='temp_1.0_max_noops_8_unclipped')
    plt.title(index)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left')
plt.tight_layout()
plt.show()
```
