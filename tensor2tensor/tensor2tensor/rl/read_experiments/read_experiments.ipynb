```python
import os
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
from collections import defaultdict
import numpy as np
import pandas as pd
```

```python
path = '/home/student/t2t_train/test1/102'
dir_data = [data for data in os.listdir(path) if data.startswith('event')]
dir_data
```

```python
summary_iterators = [EventAccumulator(os.path.join(path, dname)).Reload() for dname in dir_data]
tags = summary_iterators[0].Tags()['scalars']
```

```python
out = defaultdict(list)
steps = []
```

```python
summary_iterators[1].Scalars(tags[4])
```

```python
for tag in tags:
    steps = [e.step for e in summary_iterators[0].Scalars(tag)]
    for events in zip(*[acc.Scalars(tag) for acc in summary_iterators]):
        assert len(set(e.step for e in events)) == 1
        
        out[tag].append([e.value for e in events])
```

```python
for index, tag in enumerate(tags):
    pd.DataFrame(values[index], index=steps, columns=dir_data)
```
