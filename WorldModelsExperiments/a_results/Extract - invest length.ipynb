Invest length of each npz file. Which extraction length can be implemented

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
record = '/home/student/PycharmProjects/worldmodel/WorldModelsExperiments/breakout/record'
```

# Env_name : BreakoutFrameskip-v0

```python
filelist = os.listdir(record)
len(filelist)
```

```python
obs_list = []
reward_list_max= []
reward_list_mean = []
reward_list = []
for i in range(len(filelist)):
    filename = filelist[i]
    obs_file = np.load(os.path.join(record, filename))['obs']
    reward_file = np.load(os.path.join(record, filename))['reward']
    obs_list.append(len(obs_file))
    reward_list_max.append(max(reward_file))
    reward_list_mean.append(np.mean(reward_file))
```

```python
plt.figure(figsize=(15,8))
plt.plot(obs_list)
plt.show()
```

```python
print('Mean: ',np.mean(obs_list))
print('75 Quantil: ', np.quantile(obs_list, .75))
print('Max: ', np.max(obs_list))
```

```python
plt.figure(figsize=(15,8))
plt.plot(reward_list_max)
plt.plot(reward_list_mean)
plt.legend(['Max', 'Mean'])
plt.show()
```
